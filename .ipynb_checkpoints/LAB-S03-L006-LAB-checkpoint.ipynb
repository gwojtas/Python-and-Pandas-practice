```python
import pandas as pd
import numpy as np
```

```python
kolumny=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)']
```

```python
fuel=pd.read_csv('data/fuel.csv', low_memory=False,usecols=kolumny,index_col='Vehicle ID')
```

```python
fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
```

```python
fuel.head()
```

```python
fuel.fillna(-1)
```

```python
replaceRules={'Class':'---','Fuel Type':'---','Combined MPG (FT1)':-1}
```

```python
replaceRules
```

```python
fuel.fillna(replaceRules)
```

```python
avgMPG=fuel['Combined MPG (FT1)'].mean()
```

```python
avgMPG
```

```python
fuel['Class'].fillna('?',inplace=True)
fuel['Fuel Type'].fillna('?', inplace=True)
fuel['Combined MPG (FT1)'].fillna(avgMPG, inplace=True)
```

```python
fuel.head()
```

```python
fuel['Combined MPG (FT1)'].fillna(fuel['Combined MPG (FT1)'].min(), inplace=True)
```
